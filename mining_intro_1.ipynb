```python
import matplotlib.pyplot as plt

# Assuming 'df2' is your DataFrame (replace with your actual DataFrame)
# Example DataFrame (replace with your actual DataFrame)
data = [[1, 200], [-2, 400], [0, 300], [-3, -200]]
df2 = pd.DataFrame(data, columns=['X1', 'X2'])

# Create the scatter plot with the same scale for both axes
plt.figure(figsize=(8, 6))
plt.scatter(df2['X1'], df2['X2'], c='r')

# Set the same limits for both x and y axes
min_val = min(df2['X1'].min(), df2['X2'].min())
max_val = max(df2['X1'].max(), df2['X2'].max())
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)

# Customize the plot
plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2 (Same Scale)', fontweight='bold')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
from scipy.spatial.distance import squareform, pdist # Import squareform and pdist

# Assuming 'df2' is your DataFrame (replace with your actual DataFrame)
# Example DataFrame (replace with your actual DataFrame)
data = [[1, 200], [-2, 400], [0, 300], [-3, -200]]
df2 = pd.DataFrame(data, columns=['X1', 'X2'])

distance_matrix = squareform(pdist(df2[['X1', 'X2']]))
distance_matrix
```

# Normalization (z score & min-max normalization)

```python
# prompt: calculate z-score for df2

from scipy.stats import zscore

# Assuming df2 is defined from the previous code block
# Example DataFrame (replace with your actual DataFrame)
data = [[1, 200], [-2, 400], [0, 300], [-3, -200]]
df2 = pd.DataFrame(data, columns=['X1', 'X2'])

# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df2' is your DataFrame (replace with your actual DataFrame)
# Example DataFrame (replace with your actual DataFrame)
data = [[1, 200], [-2, 400], [0, 300], [-3, -200]]
df2 = pd.DataFrame(data, columns=['X1', 'X2'])

# Perform min-max scaling to create df2_min_max
# Import the necessary function for min-max scaling
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to your data and transform it
df2_min_max = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)


# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.plot([-0.5, 1.5], [0, 0], color='black', linewidth=2)
plt.plot([0, 0], [-0.5, 1.5], color='black', linewidth=2)
plt.scatter(df2_min_max['X1'], df2_min_max['X2'], c='r')

min_val = min(df2_min_max['X1'].min(), df2_min_max['X1'].min())
```
